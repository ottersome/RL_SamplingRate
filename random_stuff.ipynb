# Just Plot these Functions and see how min/max change over time.

```python
p00 = lambda d,mu,lam : (mu/(lam+mu)) + (lam/(lam+mu))*np.exp(-d*(lam+mu))
p01 = lambda d,mu,lam : (lam/(lam+mu)) + (lam/(lam+mu))*np.exp(-d*(lam+mu))

rates0 = {"lam": 4/10,"mu":12/10}
rates1 = {"lam": 168/10,"mu":248/10}

samp_rates = np.logspace(-3,5,1000, base=2)

fig,axs = plt.subplots(1,2)
fig.set_figwidth(12)
fig.set_figheight(4)
axs[0].plot(samp_rates, p00(samp_rates,rates0['mu'],rates0['lam']), label='p00 H0')
axs[0].plot(samp_rates, p00(samp_rates,rates1['mu'],rates1['lam']), label='p00 H1')

axs[1].plot(samp_rates, p01(samp_rates,rates0['mu'],rates0['lam']), label='p01 H0')
axs[1].plot(samp_rates, p01(samp_rates,rates1['mu'],rates1['lam']), label='p01 H1')
plt.xscale('log',base=2)
plt.show()
```
